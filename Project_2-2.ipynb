```python
print('Missing values per column:')
print(df.isna().sum())

print('\nPercentage of missing values per column:')
print((df.isna().mean() * 100).round(2))

print('\nNumber of unique values per column:')
print(df.nunique())
```

Identified missing data and assessed the number of unique values in each column to better understand the dataset structure.

```python
# 1. Clean column names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
      .str.replace('-', '_')
)
df.columns
```

Normalized column names for uniformity, improving clarity and analytical accuracy

```python
# Inspect unique values in A1CResult
df['a1cresult'].unique()[:5]
```

Identified the mode values for both a1cresult and max_glu_serum to impute the missing entries

```python
# Mode imputation for a1cresult and max_glu_serum (since they are categorical)
a1cresult_mode = df['a1cresult'].mode()[0]
max_glu_serum_mode = df['max_glu_serum'].mode()[0]
df['a1cresult'] = df['a1cresult'].fillna(a1cresult_mode)
df['max_glu_serum'] = df['max_glu_serum'].fillna(max_glu_serum_mode)
df.isna().sum()
```

Verified that no additional missing values remain

```python
# Check missing values again
df.isna().sum()
```

Converting columns to categorical data type

```python
categorical_cols = [
    'race', 'gender', 'citoglipton', 'insulin',
       'glyburide_metformin', 'glipizide_metformin',
       'glimepiride_pioglitazone', 'metformin_rosiglitazone',
       'metformin_pioglitazone', 'change', 'diabetesmed' ]

for col in categorical_cols:
    df[col] = df[col].astype('category')

df[categorical_cols].dtypes
```

This code defines a function that accepts a column name as input. It calculates two metrics for the specified column: the count of each unique value and its corresponding percentage. These results are then combined into a Pandas DataFrame and displayed.


```python
def show_category_distribution(col_name):
    counts = df[col_name].value_counts()
    percents = df[col_name].value_counts(normalize=True) * 100
    display(pd.DataFrame({'count': counts, 'percent': percents.round(2)}))

for col in ['gender', 'race', 'insulin']:
    print(f'--- {col} ---')
    show_category_distribution(col)
    print()
```

This code generates visual representations of the dataset, enabling us to uncover key insights. The charts reveal that females appear more prone to diabetes compared to males. Among ethnic groups, Caucasians show the highest susceptibility, followed by African Americans. Additionally, a significant portion of individuals are not on insulin therapy; for those who are, most maintain a steady dosage. These visualizations provide a clear understanding of demographic trends and treatment patterns within the data

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.show()

plt.figure(figsize=(8, 4))
sns.countplot(x='race', data=df)
plt.title('Race Distribution')
plt.xticks(rotation=30)
plt.show()

plt.figure(figsize=(8, 4))
sns.countplot(x='insulin', data=df)
plt.title('Insulin Distribution')
plt.xticks(rotation=30)
plt.show()
```
